```python
import numpy as np;

import matplotlib.pyplot as plt;
```

```python
x = np.arange(0, 10, 1)
x
```

```python
y = x * 2
y
```

```python
plt.plot(x, y)
plt.title("X vs Y");
plt.xlabel("X Values");
plt.ylabel("Y Values");
plt.xlim(min(x), max(x) + 0.1*max(x));
plt.ylim(min(y), max(y) + 0.1*max(y));
```

```python
fig = plt.figure();

ax = fig.add_axes([0, 0, 1, 1]);

ax.plot(x,y, label = "X vs X", color = "g");

y_sq = x ** 2;

ax.plot(x, y_sq, label = "X vs X^2", color = "r");

ax.legend()
```
